# Sea Levels Around the US (by year)

```python
import pandas as pd

filtered_sea_level = pd.read_csv("../data/clean_sea_df2.csv")
```

```python
filtered_sea_level
```

```python
sea_level_df = filtered_sea_level.sort_values(by = ['Measure', 'Date'])
sea_level_df
```

```python
sea_level_df2 = sea_level_df.drop(['Unnamed: 0', 'CTS_Name'], axis=1)
sea_level_df2
```

```python
sea_level_df2.rename(columns={'Value':'Change in MSL Value'}, inplace=True)
sea_level_df2
```

```python
sea_level_df2.insert(2, 'Year', "")
```

```python
sea_level_df2
```

```python
import datetime as dt

sea_level_df2['Date'] = pd.to_datetime(sea_level_df2['Date'])
sea_level_df2['Year'] = sea_level_df2['Date'].dt.year

sea_level_df2
```

```python
sea_level_df_grouped = sea_level_df2.groupby(['Year', 'Measure'])['Change in MSL Value'].mean()
```

```python
pd.DataFrame(sea_level_df_grouped)
```

```python
sea_level_df2.to_csv("../data/sea_level_years.csv", encoding="utf8", header=True, index=False)
```
