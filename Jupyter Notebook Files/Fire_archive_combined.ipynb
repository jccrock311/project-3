## Merging archive dataframes

```python
import pandas as pd
```

```python
# Import our datasets
first_df = pd.read_csv("../data/Fire_archive 2010-2018.csv")
second_df = pd.read_csv("../data/Fire_archive 2023.csv")
```

```python
first_df
```

```python
second_df
```

```python
first_df2 = first_df.drop('type', axis=1)
first_df2
```

```python
first_df2.dtypes
```

```python
second_df.dtypes
```

```python
# change dtype
second_df['version'] = pd.to_numeric(second_df['version'], errors = 'coerce')
```

```python
second_df.dtypes
```

```python
# Combine the data into a single dataset.
merged_df = pd.merge(first_df2, second_df, how="outer")
merged_df
```

```python
merged_df.to_csv('../data/merged_fire_archive.csv', index = True) 
```
