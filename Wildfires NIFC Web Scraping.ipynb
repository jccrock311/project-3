```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
import requests as re
```

```python
browser = Browser("chrome")
```

```python
url = "https://www.nifc.gov/fire-information/statistics/wildfires"

browser.visit(url)
```

```python
_object = browser.html

beautiful_object = soup(_object, "html.parser")
```

```python
## Finding trs which have tds with the data

rows = beautiful_object.find_all("tr") 

print(rows)
```

```python
## An empty list

empty_list = []


# Looping through the data to create a list of info 


for x in rows:
    
    table = x.find_all("td")
    
    row = [col.text for col in table] 
    
    empty_list.append(row)

    ##empty_list
    
    
    # Removing unnecessary rows with text in them
    
    
Wild_df = pd.DataFrame(empty_list, columns = ["Year", "Fires", "Acres"])

Wild_one = Wild_df.drop(0)

Wild_two = Wild_one.drop(1)

Wild_three = Wild_two.drop(2)

Wild_four = Wild_three.drop(43)

Wild_four
```
